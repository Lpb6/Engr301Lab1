*In which features are the mean and median the same in which are they significantly different (more 
than 10%)?*

```python
def percent_diff(x,y):
    percent = abs(x - y) / ((x + y)/2) * 100
    return percent
```

```python
percent_diff(wind_speed_mean, wind_speed_median)
```

```python
percent_diff(power_curve_mean, power_curve_median)
```

```python
def find_month(x):
    if " 01 " in x:
        return "Jan"
    elif " 02 " in x:
        return "Feb"
    elif " 03 " in x:
        return "March"    
    elif " 04 " in x:
        return "April"    
    elif " 05 " in x:
        return "May"    
    elif " 06 " in x:
        return "June"    
    elif " 07 " in x:
        return "July"    
    elif " 08 " in x:
        return "August"    
    elif " 09 " in x:
        return "Sep"    
    elif " 10 " in x:
        return "Oct"    
    elif " 11 " in x:
        return "Nov"    
    else:
        return "Dec"
```

```python
df["Month"]=df["Date/Time"].apply(find_month)
```

```python
df.groupby('Month').mean()
```

```python
df['Loss'] = df['Theoretical_Power_Curve (KWh)'] - df['LV ActivePower (kW)']
```

```python
df['Loss Percentage'] = df['Loss'] / df['Theoretical_Power_Curve (KWh)'] * 100
```

```python
df.head()
```

*What do these two new columns represent in terms of power produced by the wind turbine?*

> These columns show the disparity between the calculated theoretical power being produced and the actual power that is being generated by the turbine. 

```python
plt.scatter(df['Wind Speed (m/s)'],df['Loss'])
plt.title("Loss vs Wind Speed (m/s)")
plt.xlabel("Wind Speed (m/s)")
plt.ylabel("Loss (kW)")
```

```python
plt.scatter(df['Wind Speed (m/s)'],df['Loss Percentage'])
plt.title("Loss Percentage vs Wind Speed (m/s)")
plt.xlabel("Wind Speed (m/s)")
plt.ylabel("Loss Percentage (%)")
```

*How does the wind speed affect the loss?*

> The loss is most significant at lower wind speeds ( less than 20 m/s). This is seen more clearly in the Wind Speed vs Loss percentage graph where the scatter plot points lie around 0 when wind speed is over 20 m/s. With higher wind speeds, there is more loss as more power is generated. 

*What do you think is creating this loss? How could you overcome it?*

> The loss could be a result of the wind direction. This is seen in the graph of power generated versus wind direction where there are fewer points with high power generation when the wind direction is not alligned with the turbine. When the wind is approaching the turbine from the bottom or the top, even with high wind speeds, the power generation is limited. One way to overcome this would be to change the orientation of the turbine so that the wind direction is always in line with the turbine to minimize loss.
